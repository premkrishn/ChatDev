```python
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import numpy as np

# Data for the main pie chart
main_labels = ['A', 'B', 'C', 'D', 'E']
main_sizes = [40, 30, 20, 5, 5]
main_colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']

# Data for the detailed breakdown of one slice (e.g., 'E')
sub_labels = ['C1', 'C2', 'C3']
sub_sizes = [2, 2, 1]
sub_colors = ['pink', 'orange', 'purple']

# Plot the main pie chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Main pie chart
wedges, texts, autotexts = ax1.pie(main_sizes, labels=main_labels, colors=main_colors, autopct='%1.1f%%', startangle=140)
ax1.axis('equal')

# Secondary pie chart
wedges2, texts2, autotexts2 = ax2.pie(sub_sizes, labels=sub_labels, colors=sub_colors, autopct='%1.1f%%', startangle=40)
ax2.axis('equal')

# Connect the 'E' slice in main pie to the secondary pie chart
# Get the center and radius of the main pie slice
center_main = wedges[-1].center
theta1, theta2 = wedges[-1].theta1, wedges[-1].theta2
r_main = wedges[-1].r

# Calculate the midpoint angle of the main pie slice
theta_mid = (theta1 + theta2) / 2.0
x_main = r_main * np.cos(np.deg2rad(theta_mid))
y_main = r_main * np.sin(np.deg2rad(theta_mid))

# Center of the secondary pie chart (adjust this as needed)
center_sub = (ax2.get_position().x0 + ax2.get_position().width / 2.0,
              ax2.get_position().y0 + ax2.get_position().height / 2.0)

# Create the connection line
con = ConnectionPatch(xyA=(x_main, y_main), coordsA=ax1.transData,
                      xyB=(0, 0), coordsB=ax2.transData, arrowstyle="-")

# Add the connection line to the figure
fig.add_artist(con)

plt.show()
```
